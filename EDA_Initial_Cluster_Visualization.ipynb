```python
CHICAGO_COORDINATES = (41.864073, -87.706819)
# for speed purposes
MAX_RECORDS = 100

# create empty map zoomed in on San Francisco
map = folium.Map(location=CHICAGO_COORDINATES, zoom_start=11)
```

```python
def colormap(clusterIn):
    return {1:'red',2:'green'}[clusterIn]
```

```python
# add a marker for every record in the filtered data
for idx,row in init_cluster_ll.iterrows():
    folium.Circle(
        location = (row['lat'],row['lng']),
        radius = 2,
        color = colormap(row['clusterIn'])
        ).add_to(map)
#map.add_child(colormap)
display(map)
```
